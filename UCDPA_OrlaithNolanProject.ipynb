```python
import pandas as pd
import numpy as np
df = pd.DataFrame.from_dict(
{ 
    'Candidate': ['John', 'Amy', 'Lorna', 'Mark'],
    'Age': [33,44,46,38],
    'Gender': ['m', 'f', 'f','m'],
    'Job': ['developer', 'coder', 'python', 'it']
}
)
print (df)
```

```python
df['Age category'] = '40+ Years old'
```

```python
df.loc[df['Age'] < 40, 'Age Category'] = '<40 years old'
```

```python
print(df)
```

```python
conditions = [
(df['Age'] <30),
(df['Age'] <= 30) & (df['Age'] <40),
(df['Age'] <= 40) & (df['Age'] <50),
(df['Age'] <= 50) & (df['Age'] <60),
(df['Age'] <= 60),
]
```

```python
values = ['<30 years old', '30-39 years old', '40-49 years old', '50-59 years old', '60+years old' ]
```

```python
df['Age Group'] = np.select (conditions, values)
```
